## 05) escape code
code | desc
:- | :-
`\n` | 문자열 안에서 줄을 바꿀 때 사용
`\t` | 문자열 사이에 탭 간격을 줄 때 사용
`\\` | 문자 \를 그대로 표현할 때 사용
`\'` | 작은따옴표(`'`)를 그대로 표현할 때 사용
`\"` | 큰따옴표(`"`)를 그대로 표현할 때 사용
`\r` | 캐리지 리턴(줄 바꿈 문자, 현재 커서를 가장 앞으로 이동)
`\f` | 폼 피드(줄 바꿈 문자, 현재 커서를 다음 줄로 이동)
`\a` | 벨 소리(출력할 때 PC 스피커에서 `삑` 소리가 난다)
`\b` | 백 스페이스
`\000` | 널 문자

### raw string (r)
* escape code에 영향받지 않는 string
* escape code가 동작하지 않고 그대로 출력된다.
* 문자열 앞에 `r`을 붙여준다.
* json, html, windows의 path, 정규식 등에 이용

## 들여쓰기(indent)
* 다른 프로그램 언어에서 들여쓰기는 가독성을 위한 것이다.
* 파이썬은 문법적인 강제사항이다.
  * 위반 시에는 `IndentationError` 라는 에러 발생
* code block을 구성하기 위해, `if`, `for`, `class`, `def` 등등 을 작성하면서 나오는 `:` 다음 아랫줄은 반드시 들여쓰기를 해야 한다.
* 들여쓰기의 방법
  * space, tab
  * 같은 블록 내에서는 들여쓰기 칸 수가 같아야 한다.
  * 공백과 탭을 섞어쓰면 안된다.
  * 권장사항: `space 4개` ([PEP 8 -- Style Guide for Python Code](https://www.python.org/dev/peps/pep-0008/))

````
def factorial(x):
    if x == 0:
        return 1
    else:
        return x * factorial(x - 1)
````

## semi-colon(`;`)
* 다른 프로그램 언어에서는 문장의 뒤에 구문의 종료를 알리는 semi-colon(`;`)을 적는다.
* pyhon에서는 semi-colon(`;`)이 불필요하다. 줄이 바뀌면 구문이 종료된 것이다.
* semi-colon(`;`)을 붙여도 에러는 나지 않는다.
* 여러 구문을 이어 쓸때는 semi-colon(`;`)을 쓰기도 한다.

````
# 세미콜론(semicolon) 사용법
print("Hello Python!!")
print("Hello Python!!");
# print("첫번째 줄!!") print("두번째 줄") # 에러발생
print("첫번째 줄!!"); print("두번째 줄")
print("첫번째 줄!!"); print("두번째 줄");
````

## backslash(`\`), ()
* 1줄의 내용이 길어서 여러 줄로 적어야 할 경우에는 backslash(`\`) 나  ()를 이용한다.

````
# 긴 내용을 한 줄인것 처럼 인식 시키기(\ 사용)
a = 1 + 2 + 3 + \
      4 + 5 + 6
# 긴 내용을 한줄인것 처럼 인식 시키기(괄호 사용)
b = (1 + 2 + 3 +
        4 + 5 + 6)
# 파라메터를 넘기는 것이라면 콤마(comma) 뒤에서 그냥 엔터를 쓰면 됩니다.
print("Hello",  "Python",
          end="\n", sep="\t")

# 문자열 중간에 ""가 열고 닫히며 줄바꿈을 하면 1줄로 인식한다.
print("아주 아주 아주 긴 문자열이라고 가정하자 " 
         "한줄에 모두 타이핑이 힘들다면")
````

````
# 모듈(Module) 사용하기
import datetime
import math
import math as m  # module에 alias를 붙임
from math import pi # math module 중에 pi 만
from math import *  # math module 모두

# import datetime
now = datetime.datetime.now()
print(now)

# import math
print("The value of pi is", math.pi)

# import math as m
print("The value of pi is", m.pi)

# from math import pi
print("The value of pi is", pi)

# from math import *
print("The value of e is", e)
````
